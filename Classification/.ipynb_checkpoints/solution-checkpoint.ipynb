```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
def calculateMetrix(name, trueValues, predictValues):

    print(name)
    print('Confusion_matrix:')
    print(pd.DataFrame(confusion_matrix(trueValues, predictValues), index=model.classes_, columns=model.classes_))

    print('Accuracy_score:')
    print(accuracy_score(trueValues, predictValues, normalize=False))
    print('Accuracy_score_normalize:')
    print(accuracy_score(trueValues, predictValues, normalize=True))

    print('Precision_score:')
    print(precision_score(trueValues, predictValues, average=None))

    print('Recall_score:')
    print(recall_score(trueValues, predictValues,average=None))

    print('Classification_report:')
    print(classification_report(trueValues, predictValues))
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, train_size=0.7, stratify=Y)
```

```python
model = DecisionTreeClassifier(criterion='gini', min_samples_split=20, max_depth=4)
model.fit(xTrain, yTrain)
```

```python
print('#################################')
print('Classes:')
print('#################################')
print(model.classes_)
print(pd.Series(model.feature_importances_, index=X.columns))
print('#################################')
print('Model Tree:')
print('#################################')
print(model.tree_)
print(pd.DataFrame(model.predict_proba(xTest), index=xTest.index, columns=['proba_' + x for x in model.classes_]))
```

```python
yPredictTrain = model.predict(xTrain)
# print(yPredictTrain)
print('#################################')
calculateMetrix('Trening: ', yTrain, yPredictTrain)
print('#################################')
yPredictTest = model.predict(xTest)
print(yPredictTest)
print('#################################')
calculateMetrix('Test: ', yTest, yPredictTest)
```
