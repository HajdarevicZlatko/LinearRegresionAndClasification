```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data.plot(x='temp', y='count', style='o')  
plt.title('Temp vs Count')  
plt.xlabel('Temperature')  
plt.ylabel('Count')  
plt.show()
```

```python
X = data['temp'].values.reshape(-1, 1)
Y = data['count'].values.reshape(-1, 1)
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state=0)
```

```python
model = LinearRegression()  
model.fit(X_train, Y_train)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

```python
data_compare = pd.DataFrame({'Actual': Y_test.flatten(), 'Predicted': Y_pred.flatten()})
data_compare
```

```python
plt.scatter(X_test, Y_test, color='gray')
plt.plot(X_test, Y_pred, color='red', linewidth=2)
plt.show()
```

```python
count_on_temp_25 = model.intercept_ + model.coef_*25
count_on_temp_25
```

```python
data_compare_25 = pd.DataFrame({'Temp': X_test.flatten(), 'Count_predict': Y_pred.flatten()})
data_compare_25[(data_compare_25['Temp'] >= 24.60) & (data_compare_25['Temp'] <= 25.40)]
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
